```python
import nltk
import string
import numpy as np
import pandas as pd
from nltk.corpus import wordnet
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import stopwords
```

```python
training_data=pd.read_csv('0000000000002747_training_twitter_x_y_train.csv')
testing_data=pd.read_csv('0000000000002747_test_twitter_x_test.csv')
training_data.shape, testing_data.shape
```

```python
train_text=training_data['text'].values
senti=training_data['airline_sentiment'].values
train_text,senti
```

```python
documents = []
for i in range(len(train_text)):
    documents.append((word_tokenize(train_text[i]),senti[i]))
print(documents[0])
```

```python
stops=set(stopwords.words('english'))
punc=list(string.punctuation)
stops.update(punc)
print(stops)
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
lemma=WordNetLemmatizer()
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops and w.lower().isalpha():
            pos = pos_tag([w])
            clean_word = lemma.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(text), senti) for text, senti in documents]
```

```python
print(documents[0])
```

```python
test_text=testing_data['text'].values
test_text,len(test_text)
```

```python
testdoc = []
for i in range(len(test_text)):
    testdoc.append(word_tokenize(test_text[i]))
print(testdoc[0])
print(len(testdoc))
```

```python
testdoc= [(clean_review(text)) for text in testdoc]
testdoc[0]
```

```python
allwords=[]
for i in documents:
    allwords+=i[0]
print(allwords)
print(len(allwords))
```

```python
freq=nltk.FreqDist(allwords)
common=freq.most_common(4000)
features=[i[0] for i in common]
print(features)
```

```python
def get_feature_dict(words):
    curr_features={}
    word_set=set(words)
    for w in features:
        curr_features[w]=w in word_set
    return curr_features
```

```python
traindata = [(get_feature_dict(doc),category) for doc, category in documents]
print(traindata[0])
```

```python
testdata=[get_feature_dict(doc) for doc in testdoc]
```

```python
print(testdata[0])
```

```python
from nltk import NaiveBayesClassifier
classifier=NaiveBayesClassifier.train(traindata)
pred=[]
for i in range(len(testdata)):
    pred.append(classifier.classify(testdata[i]))
pred
```

```python
len(pred)
```

```python
np.savetxt("twitterpredictions.csv",pred, fmt='%s')
```
